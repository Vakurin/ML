**Import libraries:**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

**Import dataset:**

```python
dataset = pd.read_csv('Data/Mall_Customers.csv')
dataset.head(5)
```

```python
X = dataset.iloc[:, [3, 4]].values
```

**Using the dendrogram to find optimal number of clustering:**

```python
# sch.linkage - which data we apply 
# 'ward' - it's method that tried to minimize the variance each cluster
# Dendograms - based on within-cluster variance
```

```python
import scipy.cluster.hierarchy as sch
plt.figure(figsize=(20,10)) 
dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward')) 
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distance')
plt.show()
```

```python
Image('img/Dendrogram.png', width=900, height=700)
```

**Fitting hierarchical clustering:**

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 5, affinity = 'euclidean', linkage = 'ward')
y_hc = hc.fit_predict(X)
```

```python
print("First person cluster to", y_hc[0], 'group')
print("Second person cluster to", y_hc[1], 'group')
print("Third person cluster to", y_hc[2], 'group')
```

**Visualising The Clusters:**

```python
plt.figure(figsize=(20,10)) 
plt.scatter(X[y_hc == 0, 0], X[y_hc == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_hc == 1, 0], X[y_hc == 1, 1], s = 100, c = 'green', label = 'Cluster 2')
plt.scatter(X[y_hc == 2, 0], X[y_hc == 2, 1], s = 100, c = 'blue', label = 'Cluster 3')
plt.scatter(X[y_hc == 3, 0], X[y_hc == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4')
plt.scatter(X[y_hc == 4, 0], X[y_hc == 4, 1], s = 100, c = 'magenta', label = 'Cluster 5')
plt.title('Clusters Of Cliens')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend();
```
