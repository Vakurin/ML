```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
dataset = pd.read_csv('Data/Bonus System Program.csv')
```

```python
dataset.head(10)
```

```python
#if choose only [:,1] it's create size(10, )
#but else we did 1:2 it's create matrix size (10, 1)
# make sure that X is a matrix and y is a vector
X = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, 2].values
```

```python
#Fitting Linear Regression to the dataset
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

```python
#Fitting Polynomial Regression to the dataset
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 3)
X_poly = poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
```

```python
#Visualising the Linear Regression results
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg.predict(X), color = 'blue')
plt.title('Show Why Linear Regression Don\'t Work Here')
plt.xlabel('Discount Level')
plt.ylabel('Salary')
plt.show()
```

```python
#Visualising the Polynomial Regression results
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape(len(X_grid), 1)
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, lin_reg_2.predict(poly_reg.fit_transform(X_grid)), color = 'blue')
plt.title('Show Polynomial Regression')
plt.xlabel('Discount Level')
plt.ylabel('Salary')
plt.show()
```

### We Want Add New Discount Level between 8 and 9

```python
#Prediction with Linear Regression 
print(lin_reg.predict(8.5))
```

```python
#Predicition with Polynomial Regression
print(lin_reg_2.predict(poly_reg.fit_transform(8.5)))
```
