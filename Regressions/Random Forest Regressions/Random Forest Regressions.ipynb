# Random Forest Regressions

```python
# Importing the libraries
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns; sns.set()
```

```python
from sklearn.datasets import make_blobs

X, y = make_blobs(n_samples=300, centers=4,
                  random_state=0 ,cluster_std=1.0)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='rainbow');
```

```python
# Importing the dataset
dataset = pd.read_csv('Data/Bonus System Program.csv')
X = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, 2].values
```

```python
dataset.head()
```

```python
# Fitting the Random Forest Model to the dataset
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=100, random_state=0)
regressor.fit(X, y)
```

```python
# Predicting a new result
#We change numbers of trees in plot nothing changes, but in a prediction make much better
y_pred = regressor.predict(6.5)
print(y_pred)
```

```python
# Visualising the Random Forest results (for higher resolution and smoother curve)
X_grid = np.arange(min(X), max(X), 0.001)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, regressor.predict(X_grid), color = 'blue')
plt.title('Random Forest Model')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
