# Simple Vector Regression

```python
#Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
dataset = pd.read_csv('Data/Bonus System Program.csv')
X = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, 2:3].values
```

```python
dataset.head(10)
```

```python
# Feature Scaling
# In this model we need scaling cause, SVR don't support scaling inside  
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_y = StandardScaler()
X = sc_X.fit_transform(X)
y = sc_y.fit_transform(y)
```

```python
# Fitting SVR to the dataset

from sklearn.svm import SVR
svr_rbf = SVR(kernel='rbf')
svr_rbf.fit(X, y)

svr_lin = SVR(kernel='linear')
svr_lin.fit(X, y)

svr_poly = SVR(kernel='poly')
svr_poly.fit(X, y)
```

### Predicting a new result for other kernel

```python
#This method images works with GitHub
from IPython.display import Image
Image("img/Margin.png", width=500, height=500)
```

```python
#First we need transform number to scaling
#Second, first param in transform it's array therefore we used np.array
# and final we need inverse to see real number
y_pred_rbf =  sc_y.inverse_transform(
                    svr_rbf.predict(sc_X.transform(np.array([[6.5]]))))
print(y_pred_rbf)
```

```python
y_pred_lin =  sc_y.inverse_transform(
                    svr_lin.predict(sc_X.transform(np.array([[6.5]]))))
print(y_pred_lin)
```

```python
y_pred_poly =  sc_y.inverse_transform(
                    svr_poly.predict(sc_X.transform(np.array([[6.5]]))))
print(y_pred_poly)
```

```python
# Visualising SVR results (for higher resolution and smoother curve)
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'red', label = 'Data')
plt.plot(X_grid, svr_rbf.predict(X_grid), color = 'darkorange', label = 'RBF model')
plt.plot(X_grid, svr_lin.predict(X_grid), color = 'navy', label = 'Linear model')
plt.plot(X_grid, svr_poly.predict(X_grid), color = 'c', label = 'Polynomial model')
plt.title('Support Vector Regression')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.legend()
plt.show()
```
